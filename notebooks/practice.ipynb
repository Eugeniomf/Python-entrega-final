## Carga y Exploración del Dataset

- Se carga el dataset bancario.
- El dataset contiene 550 filas y 9 columnas con información de productos bancarios de clientes.
- La columna `estado de mora` fue agregada por el profesor, no tiene sentido para todos los productos (ej: Cuenta Sueldo o Cuenta Premier), pero es interesante para los préstamos y las tarjetas de crédito.

## A corregir:
- `Fecha`: está como texto, hay que convertir a tipo fecha.
- `Fecha Registro en la App`: está como texto, hay que convertir a tipo fecha.
- `Monto USD`: esta como texto, hay que convertir a numérico.

## Nulos:
- `Monto USD` y `Vendedor` tienen nulos porque solo aplican a productos de préstamos hipotecarios, consumo y automotores.
- `Fecha Registro en la App` tiene nulos porque no todos los clientes están registrados en la App.



## Data Wrangling

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
```

```python
df['Fecha Registro en la App'] = pd.to_datetime(df['Fecha Registro en la App'], format='%d/%m/%Y')
```

```python
df.dtypes
```

```python
df['Monto USD'].dtype
```

```python
df.dtypes
```

```python
df['Monto USD'] = df['Monto USD'].astype(str).str.replace(',', '.').str.strip()
df['Monto USD'] = pd.to_numeric(df['Monto USD'], errors='coerce')
df['Monto USD'].dtype
```

```python
df.dtypes
```

```python
productos_prestamo = ['Préstamo Consumo', 'Préstamo Automotor', 'Préstamo Hipotecario']

df.loc[~df['Producto'].isin(productos_prestamo), 'estado_mora'] ='N/A'

df['estado_mora'].value_counts(dropna=False)
```

La columna `estado_mora` fue relegada a productos de préstamos (Préstamo Consumo, Préstamo Automotor, Préstamo Hipotecario) ya que solo
tiene sentido de negocio para productos crediticios. Para el resto de los productos se asignó N/A.

```python
df.loc[~df['Producto'].isin(productos_prestamo), 'Vendedor'] = 'N/A'

df['Vendedor'].value_counts(dropna=False).head(10)
```

```python
df.isnull().sum()
```

```python
df.groupby('Producto')['Monto USD'].describe()
```

```python
import matplotlib.pyplot as plt

prestamos = df[df['Producto'].isin(productos_prestamo)]
prestamos['Monto USD'].hist(bins=20)
plt.title('Distribución de Monto USD (Préstamos)')
plt.xlabel('Monto USD')
plt.ylabel('Frecuencia')
plt.show()
```
